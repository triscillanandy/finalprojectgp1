```python

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import LabelEncoder

from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_squared_error,mean_absolute_error 
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
df.head(10)
```

```python
# Drop columns except 'Education Level', 'Job Title', 'Years of Experience', and 'Salary'
df = df[['Education Level', 'Job Title', 'Years of Experience', 'Salary']]

```

```python
# Drop columns except 'Education Level', 'Job Title', 'Years of Experience', and 'Salary'
df.head()
```

```python
df.info()
```

```python

df.isnull().sum()
```

```python

df = df.dropna()
df.isnull().sum()
```

```python
df['Years of Experience'] = df['Years of Experience'].astype(int)
```

```python
df.head()
```

```python
# Checking unique value counts of Job Titles in the database

df['Job Title'].value_counts()
```

```python

# Reducing Job titles by omitting titles with less than 25 counts

job_title_count = df['Job Title'].value_counts()
job_title_edited = job_title_count[job_title_count<=25]
job_title_edited.count()
```

```python
# Omitting titles with less than 25 counts

df['Job Title'] = df['Job Title'].apply(lambda x: 'Others' if x in job_title_edited else x )
df['Job Title'].nunique()
```

```python
#Checking unique value count of Education Level

df['Education Level'].value_counts()
```

```python
# Combining repeating values of education level

df['Education Level'].replace(["Bachelor's Degree","Master's Degree","phD"],["Bachelor's","Master's","PhD"],inplace=True)
df['Education Level'].value_counts()
```

```python
# detecting the outliers in salary column using IQR method
Q1 = df.Salary.quantile(0.25) # First Quartile
Q3 = df.Salary.quantile(0.75) # Third Quartile

# Caltulation Interquartile
IQR = Q3-Q1

# Deetecting outliers lying 1.5x of IQR above and below Q1 and Q3 resp
lower = Q1-1.5*IQR
upper = Q3+1.5*IQR
```

```python
df[df.Salary>upper]
```

```python
df[df.Salary<lower]
```

```python

le_education= LabelEncoder()
df['Education Level'] = le_education.fit_transform(df['Education Level'])
```

```python
# Perform label encoding on the "job title" column
le_job = LabelEncoder()
df['Job Title'] = le_job.fit_transform(df['Job Title'])
# Perform label encoding on the "job title" column
```

```python

# Dataset conntaining all features from df
x = df.drop('Salary',axis=1)

# Series containing target variable to be predicted
y = df['Salary']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

# Define the RandomForestRegressor model with a fixed number of estimators
rfr = RandomForestRegressor(n_estimators=20, random_state=42)

# Train the model
rfr.fit(x_train, y_train)

# Make predictions on the test set
y_pred_rfr = rfr.predict(x_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred_rfr)
mae = mean_absolute_error(y_test, y_pred_rfr)
rmse = mean_squared_error(y_test, y_pred_rfr, squared=False)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
```

```python
import pickle
```

```python

# Define features for prediction
X = np.array([["Bachelor's", 'Software Engineer', 5]])
```

```python
X
```

```python


# Encode the education level
X[:, 0] = le_education.transform(X[:, 0])

# Encode the job title
X[:, 1] = le_job.transform(X[:, 1])

# Convert to integers
X = X.astype(int)
X
```

```python

y_pred = rfr.predict(X)
y_pred
```

```python
data = {"model":rfr, "le_education": le_education, "le_job": le_job}  # Create a dictionary to store the model and label encoders

# Save the trained model and label encoders into a file
with open('salary_prediction_model.pkl', 'wb') as file:
    pickle.dump(data, file)
```

```python
with open('salary_prediction_model.pkl', 'rb') as file:
    data = pickle.load(file)

regressor_loaded = data["model"]
le_education = data["le_education"]
le_job = data["le_job"]
```
